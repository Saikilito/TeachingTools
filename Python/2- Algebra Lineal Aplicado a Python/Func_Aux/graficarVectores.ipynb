# Funci√≥n para graficar vectores

```python
def graficarVectores(vecs, cols, alpha = 1):
    plt.figure()
    plt.axvline(x = 0, color="grey", zorder = 0)
    plt.axhline(y = 0, color="grey", zorder = 0)
    
    for i in range(len(vecs)):
        ## concatenamos para indicar que todos los vectores
        ## que demos, tendran como origen el vector (0,0)
        x = np.concatenate([[0,0], vecs[i]])
        ## quiver nos permite graficar un vector como una flecha
        ## ax.quiver(x_pos, y_pos, x_direct, y_direct)
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                  angles='xy', 
                  scale_units='xy',
                  scale=1,
                  color=cols[i],
                  alpha=alpha)
```
