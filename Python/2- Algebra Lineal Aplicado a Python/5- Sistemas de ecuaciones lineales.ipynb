# Sistema de ecuaciones lineales

## Sistema sin soluciones
Supongamos el siguiente sistema

$$
    \alpha(x)=\left\{
                \begin{array}{ll}
                 y=3x+5\\
                 y=-x+3\\
                 y=2x+1
                \end{array}
              \right.
$$

```python
x = np.arange(-6,6)

y_1 = 3*x+5
y_2 = -x+3
y_3 = 2*x+1
```

```python
plt.figure()

plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="gray")
plt.axhline(y=0, color="gray")
plt.show()
```

Como vemos no existe ningun punto donde las tres rectas se crucen al mismo tiempo, esto queire decir que el sistema no tiene soluciÃ³n

## Sistema con una soluciÃ³n

```python
y_2 = -x+3
y_3 = 2*x+1
```

```python
plt.figure()

plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="gray")
plt.axhline(y=0, color="gray")
plt.show()
```

Como vemos las rectas se cruzan en un Ãºnico punto, entonces podemos decir que tiene una soluciÃ³n Ãºnica

## Sistemas con infinitas soluciones

```python
y_3 = 2*x+1
```

```python
plt.figure()

plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="gray")
plt.axhline(y=0, color="gray")
plt.show()
```

Si tomamos un sistema como una Ãºnica recta como ejemplo, entonces cualquier valor va a cumplir para este sistema, por lo tanto es sistema tiene infinitas soluciones
